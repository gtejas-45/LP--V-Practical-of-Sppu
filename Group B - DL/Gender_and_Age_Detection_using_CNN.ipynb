```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
# Create dummy images (simulating faces)
X = np.random.rand(500, 64, 64, 3)

# Gender labels (0 = Female, 1 = Male)
gender = np.random.randint(0, 2, 500)

# Age labels (random age 1вЂ“80)
age = np.random.randint(1, 80, 500)

print("Dataset Created Successfully")
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense

# Input layer
input_layer = Input(shape=(64,64,3))

# CNN layers
x = Conv2D(32, (3,3), activation='relu')(input_layer)
x = MaxPooling2D(2,2)(x)

x = Conv2D(64, (3,3), activation='relu')(x)
x = MaxPooling2D(2,2)(x)

x = Flatten()(x)

# Output 1: Gender
gender_output = Dense(1, activation='sigmoid', name='gender')(x)

# Output 2: Age
age_output = Dense(1, activation='linear', name='age')(x)

# Final Model
model = Model(inputs=input_layer, outputs=[gender_output, age_output])

# Compile
model.compile(
    optimizer='adam',
    loss={'gender': 'binary_crossentropy', 'age': 'mse'},
    metrics={'gender': 'accuracy', 'age': 'mae'}
)
```

```python
history = model.fit(
    X,
    {'gender': gender, 'age': age},
    epochs=5,
    batch_size=32
)
```

```python
test_img = np.random.rand(1,64,64,3)

pred_gender, pred_age = model.predict(test_img)

print("Predicted Gender:", "Male" if pred_gender[0][0] > 0.5 else "Female")
print("Predicted Age:", int(pred_age[0][0]))
```
